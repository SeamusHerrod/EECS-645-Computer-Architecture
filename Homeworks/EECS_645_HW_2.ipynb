For example, if we put in a complicated fast floating-point unit, that takes space, and something might have to be moved farther away from the middle to accommodate it, adding an extra cycle in delay to reach that unit. The basic Amdahls law equation does not take into account this trade-off. 

a.) If  the  new  fast  floating-point  unit  speeds  up  floating-point  operations  by,  on  average,  3,  and floating-point  operations  take  35% of the original programs execution time, what is the overall speedup (ignoring the penalty to any other instructions)?

$Speedup_{enhanced} = 3$

$Fraction_{enhanced} = 0.35$

$Speedup_{overall} = \frac{1}{(1 - 0.35) + \frac{0.35}{3}}$

$Speedup_{overall} = \frac{1}{0.82}  1.22$

**Speedup overall** $= 1.22$

$Speedup_{overall} = \frac{1}{_u + \frac{_{e}}{N}}$

>$_u =$ percent of the programs execution that is NOT effected by speedup/slowdown
>
>$\frac{_e}{N} = \frac{Fraction_{efp}}{Speedup_{efp}} + \frac{Fraction_{edc}}{Speedup_{edc}}$
>
>$Fraction_{efp} = 0.35$
>
>$Speedup_{efp} = 3$
>
